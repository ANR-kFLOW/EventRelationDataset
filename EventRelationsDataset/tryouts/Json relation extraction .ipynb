```python
from nltk.corpus import wordnet
synonyms = []

def get_wordnet_syn(signals):  
  for int_sign in signals:
    for syn in wordnet.synsets(int_sign):
        for l in syn.lemmas():
            synonyms.append(l.name())
  return synonyms       
  
```

```python
syn=get_wordnet_syn(intentienality_signals)
```

```python
def get_intention_json():
  intentionality_docs=[]
  
  for sentence in set(sentences):
        for sign in list(set(intentienality_signals)):
          match = re.search(' +'+sign, sentence)
          if match:
            intentionality_docs.append([sentence,sign])
  return intentionality_docs  
```
