```python
def get_sentences_EvntCaus(myroot):  
  sentences_tagged=[]
  sen_tag=[]
  text=myroot.findall("TEXT")[0]
  events=text.findall('EVENT')
  evnts=set([event.text for event in events])
  sentences=nltk.sent_tokenize(ElementTree.tostring(text).decode())
  for sentence in sentences:
    condidate_event_pairs=[evt for evt in evnts if evt in sentence]
    #sentence='<s>'+sentence+'</s>'
    sentences_tagged.append([re.sub(r'<[^>]*>|\n','',sentence),condidate_event_pairs,sentence])
    

    #print(sentence)
    #print('000000000000')
    #print(condidate_event_pairs)
    #print('///////////////////////')
    
  return sentences_tagged
```

```python
def get_sentences_EvntCaus_restrict(myroot):  
  sentences_tagged=[]
  sen_tag=[]
  text=myroot.findall("TEXT")[0]
  events=text.findall('EVENT')
  evnts=set([event.text for event in events])
  sentences=nltk.sent_tokenize(ElementTree.tostring(text).decode())
  for sentence in sentences:
    condidate_event_pairs=[evt for evt in evnts if evt in sentence]
    sen='<s>'+re.sub('<TEXT>|</TEXT>','',sentence)+'</s>'
    
    sen= ET.fromstring(sen)
    events=[(elm.text ,elm.get('eid'))for elm in  sen.findall('EVENT')]
    sentences_tagged.append([re.sub(r'<[^>]*>|\n','',sentence),events])
    

    #print(sentence)
    #print('000000000000')
    #print(condidate_event_pairs)
    #print('///////////////////////')
    
  return sentences_tagged
```

```python
def lemm_sentence(sentence):
  
  lemmas=[ token.lemma_ for token in  nlp(sentence)]
  return ' '.join(lemmas)
```

```python
lemm_sentence("I am a good programmer for languages ")
```

```python
def lemmatize_signs(sings):
  lemmetized_intentienality_signals=[]
  for sign in sings:
    doc=nlp(sign)
    for token in doc:
      lemmetized_intentienality_signals.append(token.lemma_)
  return lemmetized_intentienality_signals
```

```python
def get_intention_tml():
  intentionality_docs=[]
  for document in os.listdir():
    mytree = ET.parse(document)
    myroot = mytree.getroot()
    sentences_tagged=get_sentences_EvntCaus_restrict(myroot)
    
    #sentences_tagged=[ElementTree.tostring(element).decode() for element in myroot[2]]
    #sentences=[re.sub(r'<[^>]*>|\n','',sen) for sen in sentences_tagged ]
    #sentences=[re.sub(r'\n','',sen) for sen in sentences]
    for sentence in sentences_tagged:
      for sign in list(set(lemmetized_intentienality_signals)):
        match = re.search(r'\b'+sign+r'\b', lemm_sentence(sentence[0]))
        if match:
          intentionality_docs.append([sentence, sign,document,re.sub('\n','',myroot.find('DOCID').text)])
    
  return intentionality_docs     
```
