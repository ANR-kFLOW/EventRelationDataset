```python
def get_sentences(myroot):
  sentences_tagged=[]
  for element in myroot.findall("TEXT"):
      

      a=[[ElementTree.tostring(elem).decode(),elem] for elem in  element.findall('s')]
  

      for i in a:
        sentence=re.sub(r'<[^>]*>|\n','',i[0])
        
        events=[(elm.text,elm.get('eid')) for elm in  i[1].findall('EVENT')]
        
        sentences_tagged.append([sentence,events])
  return sentences_tagged
  

```

```python
def lemmatize_signs(sings):
  lemmetized_intentienality_signals=[]
  for sign in sings:
    doc=nlp(sign)
    for token in doc:
      lemmetized_intentienality_signals.append(token.lemma_)
  return lemmetized_intentienality_signals
```

```python
def lemm_sentence(sentence):
  
  lemmas=[ token.lemma_ for token in  nlp(sentence)]
  return ' '.join(lemmas)
```

```python
def get_intention_tml():
  intentionality_docs=[]
  for document in os.listdir():
    mytree = ET.parse(document)
    myroot = mytree.getroot()
    sentences_tagged=get_sentences(myroot)
    
    #sentences_tagged=[ElementTree.tostring(element).decode() for element in myroot[2]]
    #sentences=[re.sub(r'<[^>]*>|\n','',sen) for sen in sentences_tagged ]
    #sentences=[re.sub(r'\n','',sen) for sen in sentences]
    for sentence in sentences_tagged:
      for sign in list(set(intention)):
        match = re.search(r'\b'+sign+r'\b', lemm_sentence(sentence[0]))
        if match:
          intentionality_docs.append([ sentence,sign,document,re.sub('\n','',myroot.find('DOCNO').text)])
    
  return intentionality_docs     
```

```python
def get_causal_tml():
  causal_docs=[]
  for document in os.listdir():
    mytree = ET.parse(document)
    myroot = mytree.getroot()
    sentences_tagged=get_sentences(myroot)
    
    for sentence in sentences_tagged:
      for signal in set(catena_causal_signals):
        match2=re.search(r'\b'+',?'+signal+r'\b',sentence[0])
        if match2:
           causal_docs.append([sentence, signal,document,re.sub('\n','',myroot.find('DOCNO').text)])
      for signal in set(cau):
        match1=re.search(r'\b'+',?'+signal+r'\b',lemm_sentence(sentence[0]))
        if match1:
           causal_docs.append([sentence, signal,document,re.sub('\n','',myroot.find('DOCNO').text)])
  return  causal_docs
```
