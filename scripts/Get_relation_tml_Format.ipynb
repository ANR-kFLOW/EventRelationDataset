```python
intentienality_signals=['want','plan','aim',  'bourne' , 'bourn', 'design', 'dream', 'end', 'goal', 'idea', 'ideal', 'intent', 'mark', 'mean', 'object', 'objective', 'point', 'pretension', 'purpose', 'target']
```

```python
synonyms = []

def get_wordnet_syn(signals):  
  for int_sign in signals:
    for syn in wordnet.synsets(int_sign):
        for l in syn.lemmas():
            synonyms.append(l.name())
  return synonyms       
  
```

```python
def get_intention_tml():
  intentionality_docs=[]
  for document in os.listdir():
    mytree = ET.parse(document)
    myroot = mytree.getroot()
    sentences_tagged=[ElementTree.tostring(element).decode() for element in myroot.findall("TEXT")]
    #sentences_tagged=[ElementTree.tostring(element).decode() for element in myroot[2]]
    sentences=[re.sub(r'<[^>]*>','',sen) for sen in sentences_tagged ]
    sentences=[re.sub(r'\n','',sen) for sen in sentences]
    for sentence in sentences:
      for sign in list(set(intentienality_signals)):
        match = re.search(' +'+sign, sentence)
        if match:
          intentionality_docs.append([sentence, match,myroot[0].text,document])
  return intentionality_docs       
```
